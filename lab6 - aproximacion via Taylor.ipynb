# Aproximando Rosenbrock con Taylor

```python
from functions import rosenbrock
import numpy as np
import matplotlib.pyplot as plt
from derivatives import gradient, hessian
```

```python
f = rosenbrock

m = 50 # Density of axes
x_axis = np.linspace(-2, 2, m)
y_axis = np.linspace(-2, 2, m)
F = np.zeros((m,m))

for i in range(m):
    x0 = x_axis[i]
    for j in range(m):
        y0 = y_axis[j]
        w = np.array([x0, y0])
        F[i, j] = f(w)

[X, Y] = np.meshgrid(x_axis, y_axis)
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X,Y,F, cmap='viridis')
```

```python
def m_(x, x_hat, f):
    p = x - x_hat
    g = gradient(f, x_hat)
    B = hessian(f, x_hat)
    return 1/2*(p.T@B@p) + (g.T@p) + f(x_hat)
```

```python

m = 50 # Density of axes
x_axis = np.linspace(0, 2, m)
y_axis = np.linspace(0, 2, m)
F = np.zeros((m,m))

x_hat = np.array([1.,1.])

for i in range(m):
    x0 = x_axis[i]
    for j in range(m):
        y0 = y_axis[j]
        w = np.array([x0, y0])
        F[i, j] = m_(w, x_hat, rosenbrock)

[X, Y] = np.meshgrid(x_axis, y_axis)
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X,Y,F, cmap='viridis')
```
