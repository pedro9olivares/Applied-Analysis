```python
from derivatives import gradient, hessian
import numpy as np
```

```python
def mi_gc(A, b, maxiterk=200):
    k = 0
    tol = 10**-8
    n = len(b)
    x = np.zeros(n)

    r = np.dot(A,x) - b
    p = -r

    while np.linalg.norm(r) > 0 and k < maxiterk:
        alpha = - (r.T@p)/(p.T@A@p)
        x = x + alpha*p
        r = np.dot(A,x) - b

        v = np.dot(A, p)
        den = np.inner(p,v)
        beta = np.inner(r, v)/den

        p = -r + beta*p

        k+=1
        print(f'iteración {k} = {np.linalg.norm(r)}')

    return x, k
```

```python
# Probando
A = np.zeros((50,50))

for i in range(49):
    A[i,i] = 3
    A[i,i+1] = 1
    A[i+1,i] = 1
A[49,49] = 3

b = np.ones(50)
```
